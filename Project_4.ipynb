```r
#1
glimpse(Fingers)
```

```r
#2
ggplot(data = Fingers, mapping = aes(x = Thumb)) +
geom_histogram()
```

```r
#3
gf_histogram(~ Thumb, data = Fingers, color = ~Sex) %>%
  gf_refine(scale_color_manual(values = c("purple", "orange")))
```

```r
#3
gf_point(Thumb ~ Sex, data = Fingers)
```

```r
#4
gf_jitter(Thumb ~ Sex, data = Fingers)
```

5. The benefit of using jitter plot is to see the points clearly. Jitter plot make the points do not overlap each other, horizontal difference does not matter because there is only female and male in this case in x axis.  


```r
#6
gf_jitter(Thumb ~ Sex, data = Fingers, alpha = 0.2)
gf_jitter(Thumb ~ Sex, data = Fingers, alpha = 0.5)
gf_jitter(Thumb ~ Sex, data = Fingers, alpha = 1)
```

6. Alpha changes the transparency of the plots.
