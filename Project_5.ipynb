```r
#3
ggplot(evals, aes(bty_avg, score)) + geom_jitter()
```

3. Jitter means to add noise to the vector. Since many points were overlapping on the other ones, we were able to see little bit in the original scatter plot, but by using the jitter function, we can see the overlapping points better because the points were jittered little.

```r
#4
score_bty_fit <- lm(score ~ bty_avg, data = evals)
plot(jitter(evals$score) ~ jitter(evals$bty_avg))
abline(score_bty_fit)
tidy(score_bty_fit)
summary(score_bty_fit)
```

4. By doing the calculation, the slope is about $\frac{4.4-4}{8.2-1.8}=0.0625$. Then, by plugging in $(2,4)$ into score = 0.0625(bty_avg) +b, we find $3.875$ as $b$. Therefore the slope is score = 0.0625(bty_avg) +3.875$

```r
#5
score_bty_fit2 <- lm(score ~ bty_avg, data = evals)
plot((evals$score) ~ (evals$bty_avg))
abline(score_bty_fit2)
tidy(score_bty_fit2)
```

6. This slope has a positive correlation between beauty average and score.

score = 0.06663704(bty_avg) + 3.88033795



7. When the beauty average is 0, the score would be 3.875 according to the slope function that we calulated. When the score is 0, bty_avg will have a negative number, so it does not make sense.

```r
#8
glance(score_bty_fit2)$r.squared
```
