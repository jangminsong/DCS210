```r
#4 1)
no_na_edibnb <- na.omit(edibnb)
no_na_edibnb %>%
  group_by(neighbourhood) %>%
  summarise(median_price = median(price)) %>%
  arrange(-median_price) %>%
  slice(1:5)
```

```r
#4 2)
top5 <- na.omit(edibnb)%>%
    filter(neighbourhood == "New Town" | neighbourhood == "Old Town" |neighbourhood == "West End" |
        neighbourhood == "Stockbridge" | neighbourhood == "Southside") %>%
    select(neighbourhood, price)
```

```r
#4 2)
ggplot(top5, aes(x= price, y = neighbourhood)) +
    geom_density_ridges() +
    theme_ridges() +
    theme(legend.position = "none")
```

```r
#4 3)
top5 %>%
  group_by(neighbourhood) %>%
  summarise(
    min_price = min(price),
    mean_price = mean(price),
    median_price = median(price),
    sd_price = sd(price),
    IQR_price = IQR(price),
    max_price = max(price)
  )
```
